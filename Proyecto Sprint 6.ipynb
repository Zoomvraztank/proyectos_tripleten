**¿Cuánto dura una plataforma?**

Para aproximar cuánto tardan en aparecer y desaparecer, calculamos el primer y último año donde cada plataforma registra ventas (mayores a 0).
Esto es una aproximación descriptiva basada en el dataset.


```python

# Ventas por año y plataforma (todo el dataset)
all_platform_year_sales = df.pivot_table(
    index="year_of_release",
    columns="platform",
    values="total_sales",
    aggfunc="sum"
)

lifespan = []

for platform in all_platform_year_sales.columns:
    s = all_platform_year_sales[platform].dropna()
    s = s[s > 0]  # Se consideran años con ventas > 0

    if len(s) > 0:
        first_year = int(s.index.min())
        last_year = int(s.index.max())
        lifespan_years = last_year - first_year

        lifespan.append([platform, first_year, last_year, lifespan_years])

lifespan_df = pd.DataFrame(
    lifespan,
    columns=["platform", "first_year", "last_year", "lifespan_years"]
)

lifespan_df.sort_values("lifespan_years", ascending=False)

```

```python
lifespan_df.boxplot(column="lifespan_years",figsize=(6,4))
plt.title("Distribución del ciclo de vida de las plataformas")
plt.ylabel("Años de vida aproximados")
plt.show()
```

Existen valores atípicos que representan plataformas con presencia prolongada a lo largo del tiempo, como PC, que no siguen el ciclo típico de una consola. Por ello, no se consideran representativas del ciclo estándar de vida de una plataforma.

A partir del boxplot del ciclo de vida de las plataformas se observa que la mediana de duración es de aproximadamente 6 años.
La mayoría de las plataformas se concentra en un rango aproximado entre 3 y 10 años.

También aparecen valores atípicos con duraciones mucho mayores que corresponden a plataformas como PC o categorías similares, las cuales no siguen un ciclo de vida típico y permanecen activas durante muchos años.

Debido a esto el ciclo de vida promedio de una plataforma puede estimarse en alrededor de 6 años.
Con base en esta observación y considerando que el análisis se sitúa en 2016, resulta razonable enfocarse en un periodo reciente que represente el estado actual del mercado.

Por esta razón, se selecciona un rango cercano a los últimos años (2013–2016), el cual captura plataformas activas y evita que los resultados estén dominados por generaciones antiguas.

**COMPARACION DE VENTAS DEL MISMO JUEGO EN DISTINTAS PLATAFORMAS**

Algunos videojuegos se lanzan en varias plataformas al mismo tiempo. Para analizar si la plataforma influye en las ventas comparamos las ventas globales de los mismos juegos cuando aparecen en más de una plataforma.

Para ello se identifican los juegos que aparecen repetidos (mismo nombre) y se analizan sus ventas según la plataforma en la que fueron lanzados.


```python
#Contamos cuántas veces aparece cada juego
game_counts=df_recent["name"].value_counts()

#Seleccionamos juegos que aparecen en más de una plataforma
multi_platform_games=game_counts[game_counts>1].index

#Filtramos solo esos juegos
multi_df=df_recent[df_recent["name"].isin(multi_platform_games)]
```

```python
comparison_table=multi_df.pivot_table(
    index="name",
    columns="platform",
    values="total_sales",
    aggfunc="sum"
)

comparison_table.head()
```

```python
#Contamos cuantas plataformas reales tiene cada juego
platform_counts = comparison_table.notna().sum(axis=1)
```

```python
multi_platform_only = comparison_table[platform_counts > 1] #Filtramos solo juegos con más de una plataforma
```

```python
multi_platform_only.head(10)
```

Para comparar correctamente las ventas de un mismo juego en distintas plataformas es necesario asegurarse de que el juego tenga ventas registradas en más de una plataforma.

Para ello se contó cuántas plataformas presentan valores no nulos de ventas para cada juego. Posteriormente se conservaron unicamente aquellos juegos con ventas en dos o más plataformas.

De esta forma se evita considerar juegos que aparecen en una sola plataforma y se garantiza una comparación valida entre plataformas.


```python
# Vamos a gráficar algunos juegos para que no sea ilegible

top_multi_games = (
    multi_platform_only.sum(axis=1)
    .sort_values(ascending=False)
    .head(10)
    .index
)

top_multi_games
```

```python
comparison_top = multi_platform_only.loc[top_multi_games]
comparison_top
```

```python
comparison_top.plot(
    kind="bar",
    figsize=(12, 6)
)

plt.title("Comparación de ventas de los mismos juegos en distintas plataformas")
plt.xlabel("Juego")
plt.ylabel("Ventas globales (millones USD)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

El gráfico muestra las ventas de los mismos juegos cuando están disponibles en diferentes plataformas.

Se observa que un mismo título puede tener resultados muy distintos dependiendo de la plataforma en la que se lance. En varios casos las ventas son considerablemente mayores en ciertas consolas lo que indica que la base de usuarios y la popularidad de la plataforma influyen directamente en el desempeño comercial del juego.

Este análisis refuerza la idea de que no solo el contenido del juego determina su éxito sino también la plataforma en la que se distribuye.

**PRUEBAS DE HIPOTESIS**

 Hipótesis 1: comparación entre Xbox One y PC

**Hipótesis nula (H₀):**  
La calificación promedio de los usuarios para los juegos de Xbox One es igual a la calificación promedio de los usuarios para los juegos de PC.

**Hipótesis alternativa (H₁):**  
La calificación promedio de los usuarios para los juegos de Xbox One es diferente de la calificación promedio de los juegos de PC.

Se establece un nivel de significancia α = 0.05.


```python
# Filtramos datos para Xbox One y PC
xbox_users = df[df["platform"] == "XOne"]["user_score"]
pc_users = df[df["platform"] == "PC"]["user_score"]

# Eliminamos valores faltantes
xbox_users = xbox_users.dropna()
pc_users = pc_users.dropna()
```

```python
#Usamos ttest_ind porque comparamos medios de dos muestras independientes
from scipy import stats

results_1 = stats.ttest_ind(xbox_users, pc_users, equal_var=False)
results_1
```

**Resultados de la prueba de hipótesis**

Se realizó una prueba t para muestras independientes con el objetivo de comparar las calificaciones promedio de los usuarios entre los juegos de **Xbox One** y **PC**.

- Nivel de significancia (α): 0.05  
- Estadístico t obtenido: -4.65  
- Valor p (p-value): 4.93e-06  

**Decisión estadística**

Dado que el valor p es *mucho menor que α = 0.05*, se **rechaza la hipótesis nula**.

Esto significa que existe evidencia estadísticamente significativa para afirmar que:

*La calificación promedio de los usuarios para juegos de Xbox One es diferente a la de los juegos de PC.*

**Interpretación práctica**

Aunque ambas plataformas tienen muchos títulos bien valorados las diferencias observadas en las calificaciones no se deben al azar. Esto sugiere que:

- La percepción de los usuarios varía según la plataforma.
- Factores como tipo de catálogo, rendimiento, comunidad o expectativas pueden influir en las calificaciones.
- Las calificaciones de usuarios pueden ser un indicador util pero no suficiente por si solas para predecir ventas.

Este resultado se usará más adelante junto con otros análisis (ventas, regiones y géneros) para entender mejor el comportamiento del mercado.


**Prueba de hipótesis: calificaciones de usuarios en juegos de Acción vs Deportes**

Ahora analizamos si existen diferencias significativas entre las calificaciones promedio de los usuarios para los juegos de **Acción** y los juegos de **Deportes**.

**Hipótesis nula (H₀):**  
La calificación promedio de los usuarios para los juegos de Acción es igual a la calificación promedio de los juegos de Deportes.

**Hipótesis alternativa (H₁):**  
La calificación promedio de los usuarios para los juegos de Acción es diferente de la calificación promedio de los juegos de Deportes.

Se utilizará un nivel de significancia α = 0.05.


```python
# Filtramos juegos del género Acción y Deportes
action_users = df[df["genre"] == "Action"]["user_score"]
sports_users = df[df["genre"] == "Sports"]["user_score"]

# Eliminamos valores faltantes
action_users = action_users.dropna()
sports_users = sports_users.dropna()
```

```python
results_2 = stats.ttest_ind(action_users, sports_users, equal_var=False)
results_2
```

**Resultados de la prueba de hipótesis**

Se aplicó una prueba t para muestras independientes con el fin de comparar las calificaciones promedio de los usuarios entre los géneros **Acción** y **Deportes**.

- Nivel de significancia (α): 0.05  
- Valor p obtenido: 0.1148  

**Decisión estadística**

Dado que el valor p es mayor que el nivel de significancia (0.05), **no se rechaza la hipótesis nula**.

**Conclusión**

No se encontró evidencia estadísticamente significativa para afirmar que las calificaciones promedio de los usuarios difieren entre los juegos de Acción y los juegos de Deportes.

Esto indica que aunque puedan existir diferencias individuales entre títulos, en promedio ambos géneros reciben valoraciones similares por parte de los usuarios.
